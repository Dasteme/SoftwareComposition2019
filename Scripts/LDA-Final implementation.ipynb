```python
import numpy as np
import glob
import os
import csv
import nltk
from nltk.stem import PorterStemmer



csvdir = 'D:/Uni Bern/7. Semester/Software Composition/Github/SoftwareComposition2019/Stackoverflow_subposts_cleaned/*.csv'
current_document = []
current_post_link = ""
porter = PorterStemmer()




def generate_document(newPostRow):
    global current_document
    global current_post_link
    print("Generating document for: " + newPostRow[0]);
    current_post_link = newPostRow[0]
    current_document = [newPostRow[1], newPostRow[2]]; # Adding title and question
    
def add_answer(answer):
    global current_document
    current_document.append(answer);


def finish_document():
    global current_document
    global current_post_link
    if (current_post_link == ""):
        return
    
    apply_porter(current_document)
    
    print (current_document)
    

def apply_porter(array):
    i = 0
    for elem in array:
        array[i] = porter.stem(elem)
        i = i+1
    




csvs = glob.glob(csvdir)

for c in csvs:
    file = c.replace('\\', '/', 1)
    newfile = c.replace('Stackoverflow_subposts', 'Stackoverflow_subposts_documents', 1)
    
    
    # Open post file and read it
    f = open(file, encoding='utf8')
    csvreader = csv.reader(f, delimiter=',', quotechar="\"", quoting=csv.QUOTE_ALL)
    
    
    
    print("**************************************************");
    i = 0;
    for row in csvreader:
        row[1] = "title"
        row[2] = "question"
        row[3] = "answer"
        if (current_post_link == row[0]):
            add_answer(row[3]);
        else:
            finish_document()
            generate_document(row);
            add_answer(row[3]);
        
        i = i+1
        if (i > 5):
            break;
    finish_document()
    
    print("**************************************************");
    
    break;





```
